#Importing the depindences

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image
import tensorflow as tf
tf.random.set_seed(3)
#keras developed by google and pythourch developed by facebook
from tensorflow import keras
from keras.datasets import mnist
from tensorflow.math import confusion_matrix
```

#Data collection

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
# unique values in Y_train
print(np.unique(Y_train))

# unique values in Y_test
print(np.unique(Y_test))
```

we can use these labels as such or we can also apply One Hot Encoding

all the images have the same dimensions in this dataset, if not, we have to resize all the images to a common dimension

```python
# scaling the values , if we don't do this maybe we have the lower accuracy so better to normalize this by dividing it on maximum value

X_train = X_train/255
X_test = X_test/255
```

when we apply this the value will be in the range of (0,1)

beacause:

minimum --> 0/255 = 0

and

maximum --> 255/255 = 1

```python
print(X_train[10])
```

#building the Neural Network

```python
# setting up the layers for Neural Network
# frist layer its input layer second and therd layers are hidden layers and 4th layer its output layer
model = keras.Sequential([
                          keras.layers.Flatten(input_shape = (28,28)),
                          keras.layers.Dense(50, activation='relu'),
                          keras.layers.Dense(50, activation='relu'),
                          keras.layers.Dense(10, activation='sigmoid')

])
```

```python
# compiling the Neural Network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Neural Network
model.fit(X_train, Y_train, epochs=10)
```

Training data accuracy = 98.9%

**Accuracy on Test Data:**

```python
loss, accuracy = model.evaluate(X_test, Y_test)
print(accuracy)
```

Test data accuracy = 97.1%

```python
# first data point in the x test
plt.imshow(X_test[0])
```

```python
print(Y_test[0])
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred.shape)
```

```python
print(y_pred[0])
```

**model.predict() gives the prediction probability of each class for that data point**

```python
# converting the prediction probability to class label
label_for_first_test_image = np.argmax(y_pred[0])
print(label_for_first_test_image)
```

```python
# converting the prediction probability to class label for all test data point

Y_pred_labels = [np.argmax(i) for i in y_pred]
print(Y_pred_labels)
```

Y_test --> True labels

Y_pred_label --> Predicted labels

#Confusion Matrix

```python
conf_mat = confusion_matrix(Y_test, Y_pred_labels)
```

```python
print(conf_mat)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues')
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

The diagonal represnt the true predicted made by our model
